```python
import datetime
```

```python
t1 = datetime.datetime.now()
print(f'started at {t1}')

output = model.transcribe("milton_friedman_trimmed.mp4")

t2 = datetime.datetime.now()
print(f'ended at {t2}')

print(f'time elapsed: {t2-t1}')
```
